```python
from langchain_google_genai import ChatGoogleGenerativeAI
# Switching to gemini-pro to avoid resource exhaustion errors or not found errors
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash", api_key=API_KEY)
```

```python
from langchain_community.document_loaders import PyPDFLoader
mypdf = PyPDFLoader('/StayingAhead_Daily_2026_05_22.pdf')
lodded_doc = mypdf.load()
```

```python
!pip install -U langchain-text-splitters
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
recursive_splitter = RecursiveCharacterTextSplitter(chunk_size=400, chunk_overlap=40, separators=['\n\n', '\n', ' '])
```

```python
chunks = recursive_splitter.split_documents(lodded_doc)
print(len(chunks))
```

```python
!pip install chromadb
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
embeddings = GoogleGenerativeAIEmbeddings(model='models/gemini-embedding-001',api_key=API_KEY)
```

```python
from langchain_core.vectorstores import InMemoryVectorStore
vector_store = InMemoryVectorStore.from_documents(documents=chunks,embedding=embeddings)
```

```python
vector_index = vector_store.as_retriever(search_kwargs={'k':8})
```

```python
from langchain_classic.chains import RetrievalQA
rag = RetrievalQA.from_chain_type(
    model,
    retriever = vector_index,
    return_source_documents=True
)
```

```python
question =input('enter the question:')
# Using invoke for the latest LangChain standards
response = rag.invoke({"query": question})
print(response['result'])
```
